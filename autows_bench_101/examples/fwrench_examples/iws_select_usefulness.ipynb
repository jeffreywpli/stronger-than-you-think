```python
def plot_acc_cov_thresh(dataset, axs, title=''):
    with open(f"{dataset}_acc_file.txt", "r") as file:
        acc_list=file.read().splitlines()
    acc_list = list(map(float, acc_list))
    with open(f"{dataset}_cov_file.txt", "r") as file:
        cov_list=file.read().splitlines()
    cov_list = list(map(float, cov_list))
    thresholds = np.arange(0.1, 1.0, 0.1)

    axs[0].plot(thresholds, acc_list[:len(thresholds)], color='red', marker='o')
    axs[0].set_title(f'{title}')
    axs[0].set_xlabel('Accuracy Threshold')
    axs[0].set_ylabel('Accuracy')
    #axs[0].grid(True)

    axs[1].plot(thresholds, cov_list[:len(thresholds)], color='red', marker='o')
    axs[1].set_xlabel('Accuracy Threshold')
    axs[1].set_ylabel('Coverage')
    #axs[1].grid(True)





fig, axs = plt.subplots(2, 3, figsize=(10, 4), sharex=True, sharey=True)

plot_acc_cov_thresh('mnist', axs[:, 0], title='MNIST + CLIP')
plot_acc_cov_thresh('navier_stokes', axs[:, 1], title='Navier-Stokes + ResNet-18')
plot_acc_cov_thresh('yelp', axs[:, 2], title='Yelp + Raw features')

plt.tight_layout()
plt.savefig('../../neurips/iws_ablation_thresh.pdf')
```
